```python
import pandas as pd
import matplotlib.pyplot as plt
import random
from part_2 import kMeansAlgorithm 
from matplotlib import colors
```

```python
# Read the CSV file
data = pd.read_csv('kmeans.csv')

# Extract the x and y values from the data
x = data['f1']
y = data['f2']

# Create the scatterplot
plt.scatter(x, y)
plt.xlabel('f1')
plt.ylabel('f2')
plt.title('K-means')


# Display the plot
plt.show()
```

```python
k = 2
assignedData = kMeansAlgorithm(k, data['f1'], data['f2'])
```

```python
colours = list(colors.TABLEAU_COLORS.keys())
random.shuffle(colours)
colours = colours[:k]
```

```python
colourCount = 0 
for centroid, points in assignedData.items():
    colour = colours[colourCount]
    colourCount += 1

    centroid_x = centroid.x
    centroid_y = centroid.y

    # Plot centroid
    plt.scatter(centroid_x, centroid_y, marker='+', color=colour, s=100)

    # add centroid labels
    plt.text(centroid_x, centroid_y, f"({centroid_x}, {centroid_y})", fontsize=8)

    # Plot data points
    for point in points:
        plt.scatter(point.x, point.y, marker='o', color=colour)
    



plt.xlabel('f1')
plt.ylabel('f2')
plt.title('K-means Clustering Results')
plt.show()
```
