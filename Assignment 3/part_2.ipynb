```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the CSV file
data = pd.read_csv('kmeans.csv')

# Extract the x and y values from the data
x = data['f1']
y = data['f2']

# Create the scatterplot
plt.scatter(x, y)
plt.xlabel('f1')
plt.ylabel('f2')
plt.title('K-means')


# Display the plot
plt.show()
```
