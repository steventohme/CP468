```python
import csv
import matplotlib.pyplot as plt

# Read the city data from the CSV file
city_data = []
with open('city_data_50.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader)  # Skip the header row
    for row in reader:
        name, description, latitude, longitude = row
        city_data.append((description, float(latitude), float(longitude)))

# Set figure size
plt.figure(figsize=(10, 8))

# Draw the graph with circles as cities
for description, latitude, longitude in city_data:
    plt.plot(longitude, latitude, 'o', color='red', markersize=20)
    plt.text(longitude, latitude, description, fontsize=8, ha='center', va='center')

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Cities Graph')
plt.grid(True)
plt.tight_layout()
plt.show()
```
